```python
#Various denoising filters

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters, restoration
from skimage.color import rgb2gray

# Load an image
image = io.imread('Ganesh.jpg')  

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = rgb2gray(image)

# Apply Gaussian filter
gaussian_filtered_image = filters.gaussian(image, sigma=2)

# Apply Median filter
median_filtered_image = filters.median(image)

# Apply Total Variation filter
tv_filtered_image = restoration.denoise_tv_chambolle(image, weight=0.1)

# Display the original and filtered images
plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gaussian_filtered_image, cmap='gray')
plt.title('Gaussian Filtered Image')

plt.subplot(2, 2, 3)
plt.imshow(median_filtered_image, cmap='gray')
plt.title('Median Filtered Image')

plt.subplot(2, 2, 4)
plt.imshow(tv_filtered_image, cmap='gray')
plt.title('Total Variation Filtered Image')

plt.show()
```

```python
#sobel filter
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters

# Load an image
image = io.imread('Ganesh.jpg')  

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Apply Sobel filter
sobel_filtered_image = filters.sobel(image)

# Display the original and Sobel-filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(sobel_filtered_image, cmap='gray')
plt.title('Sobel Filtered Image')

plt.show()
```

```python
#Equilise histogram
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, exposure

# Load an image
image = io.imread('Ganesh.jpg')  

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Perform histogram equalization
equalized_image = exposure.equalize_hist(image)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.show()
```

```python
#INVERSE FILTERING

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, restoration
from scipy.ndimage import convolve

# Load a degraded image
degraded_image = io.imread('Ganesh.jpg')
# Convert the image to grayscale if it's a color image
if len(degraded_image.shape) == 3:
    degraded_image = color.rgb2gray(degraded_image)

# Simulate a known degradation (e.g., motion blur)
# In a real scenario, you might have the kernel from the actual degradation process
kernel = np.ones((5, 5)) / 25
degraded_image = convolve(degraded_image, kernel, mode='nearest')

# Apply inverse filtering with Tikhonov regularization
# Adjust alpha for controlling regularization strength
alpha = 0.1
restored_image = restoration.wiener(degraded_image, kernel, alpha)

# Display the original, degraded, and restored images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(degraded_image, cmap='gray')
plt.title('Degraded Image')

plt.subplot(1, 3, 2)
plt.imshow(kernel, cmap='gray')
plt.title('Degradation Kernel')

plt.subplot(1, 3, 3)
plt.imshow(restored_image, cmap='gray')
plt.title('Restored Image (Wiener Deconvolution)')

plt.show()
```

```python
#Image compression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage import io, color

# Load an image
image = io.imread('Ganesh.jpg') 

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Flatten the 2D image into a 1D array
original_shape = image.shape
flattened_image = image.flatten()

# Apply PCA for image compression
num_components = 50  # Number of principal components to keep (adjust as needed)
pca = PCA(n_components=num_components)
compressed_image = pca.fit_transform(image)

# Reconstruct the compressed image
reconstructed_image = pca.inverse_transform(compressed_image)
reconstructed_image = reconstructed_image.reshape(original_shape)

# Display the original and compressed images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Compressed Image ({} components)'.format(num_components))

plt.show()
```

```python
#Edge DEtection
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, feature

# Load an image
image = io.imread('eagle.jpg')  

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Apply the Canny edge detector
edges = feature.canny(image, sigma=1)

# Display the original and edges-detected images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges Detected')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, segmentation

# Load an image
image = io.imread('eagle.jpg')  

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Perform segmentation using Felzenszwalb's method
segments = segmentation.felzenszwalb(image, scale=100, sigma=0.5, min_size=50)

# Display the original and segmented images
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(segments, cmap='nipy_spectral')  # 'nipy_spectral' colormap for better visualization
plt.title('Segmented Image')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters

# Load an image
image = io.imread('eagle.jpg')  # Replace 'your_image.jpg' with the path to your image

# Convert the image to grayscale if it's a color image
if len(image.shape) == 3:
    image = color.rgb2gray(image)

# Compute the optimal threshold using Otsu's method
threshold_value = filters.threshold_otsu(image)

# Create a binary mask using the threshold
binary_image = image > threshold_value

# Display the original and thresholded images
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='binary')
plt.title('Thresholded Image (Otsu\'s Method)')

plt.show()
```
